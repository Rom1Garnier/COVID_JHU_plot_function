```python
import matplotlib.pyplot as plt
import seaborn as sns
from COVID_plot_function import COVID_plots
```

```python
#A single plot
plt.figure(figsize = (15, 7.5))
COVID_plots(locations = ['Arizona'], data_type = 'cases', fig_type = 'trend')
plt.xlabel('Date', size = 20)
plt.ylabel('Cumulative cases', size = 20)
plt.title('Cumulative COVID-19 cases in Arizona (June 24th)', size = 24)
```

```python
#It works with any number of states

plt.figure(figsize = (30, 7.5))

ax1 = plt.subplot(131)
ax2 = plt.subplot(132)
ax3 = plt.subplot(133)

COVID_plots(locations = ['Arizona', 'Texas', 'Florida', 'California'], data_type = 'cases', fig_type = 'cumulative', ax = ax1)
ax1.set_xlabel('Date', size = 20)
ax1.set_ylabel('Cumulative cases', size = 20)
ax1.set_title('Cumulative COVID-19 cases', size = 24)

COVID_plots(locations = ['Arizona', 'Texas', 'Florida', 'California'], data_type = 'cases', fig_type = 'new', ax = ax2)
ax2.set_xlabel('Date', size = 20)
ax2.set_ylabel('Cumulative cases', size = 20)
ax2.set_title('Cumulative COVID-19 cases', size = 24)

COVID_plots(locations = ['Arizona', 'Texas', 'Florida', 'California'], data_type = 'cases', fig_type = 'trend', ax = ax3)
ax3.set_xlabel('Date', size = 20)
ax3.set_ylabel('Trend in daily new cases', size = 20)
ax3.set_title('Trend in daily new COVID-19 cases', size = 24)
```

```python
#It can also be done with deaths data rather than cases

plt.figure(figsize = (30, 7.5))

ax1 = plt.subplot(131)
ax2 = plt.subplot(132)
ax3 = plt.subplot(133)

COVID_plots(locations = ['Virginia', 'Florida', 'California'], data_type = 'deaths', fig_type = 'cumulative', ax = ax1)
ax1.set_xlabel('Date', size = 20)
ax1.set_ylabel('Cumulative cases', size = 20)
ax1.set_title('Cumulative COVID-19 deaths', size = 24)

COVID_plots(locations = ['Virginia', 'Florida', 'California'], data_type = 'deaths', fig_type = 'new', ax = ax2)
ax2.set_xlabel('Date', size = 20)
ax2.set_ylabel('Cumulative cases', size = 20)
ax2.set_title('Cumulative COVID-19 deaths', size = 24)

COVID_plots(locations = ['Virginia', 'Florida', 'California'], data_type = 'deaths', fig_type = 'trend', ax = ax3)
ax3.set_xlabel('Date', size = 20)
ax3.set_ylabel('Trend in daily new cases', size = 20)
ax3.set_title('Trend in daily new COVID-19 deaths', size = 24)
```
